### Importing the 5 models

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'SVM': SVC(random_state=42),
    'XGBoost': XGBClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=5000,random_state=42),
    'KNN': KNeighborsClassifier()
}
```

```python
model=['Random Forest','SVM','XGBoost','Logistic Regression','KNN']
answer=pd.DataFrame({'model':model})
answer
```

### Random Sampling

```python
from sklearn.model_selection import train_test_split
x=data_df.drop('Class',axis=1)
y=data_df['Class']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sampling_1=[]
for model_name, model in models.items():
 
    model.fit(x_train, y_train)
    
   
    train_predictions = model.predict(x_test)
    train_accuracy = accuracy_score(y_test, train_predictions)
    
    
    sampling_1.append(train_accuracy)
 

```

```python
answer['Sampling1']=sampling_1
answer
```

### Systematic Sampling 

```python
import numpy as np
sampling_interval=5
start_index=np.random.randint(0,sampling_interval)
systematic_samples = data_df.iloc[start_index::sampling_interval]
data_test = data_df.drop(systematic_samples.index)
```

```python
systematic_samples.shape
```

```python
sampling_2=[]
x_train=systematic_samples.drop('Class',axis=1)
y_train=systematic_samples['Class']
x_test=data_test.drop('Class',axis=1)
y_test=data_test['Class']
for model_name, model in models.items():
 
    model.fit(x_train, y_train)
    
   
    train_predictions = model.predict(x_test)
    train_accuracy = accuracy_score(y_test, train_predictions)
    
    
    sampling_2.append(train_accuracy)
answer['Sampling2']=sampling_2
 
```

### K-Fold Cross Validation

```python
from sklearn.model_selection import KFold, cross_val_score
sampling_3=[]
k_folds=10
x=data_df.drop('Class',axis=1)
y=data_df['Class']
for model_name,model in models.items():
    kfold = KFold(n_splits=k_folds, shuffle=True, random_state=42)
    cv_results = cross_val_score(model, x, y, cv=kfold, scoring='accuracy')
    sampling_3.append(cv_results.mean())
```

```python
answer['Sampling3']=sampling_3
answer
```

### Cluster Sampling

```python
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
target_variable = 'Class'


X = data_df.drop(target_variable, axis=1)
y = data_df[target_variable]


num_clusters = 10


kmeans = KMeans(n_clusters=num_clusters, random_state=42)
data_df['Cluster'] = kmeans.fit_predict(X)

print(data_df['Cluster'].value_counts())


def clustering_sampling(clustered_df, num_samples_per_cluster):
    sampled_data = pd.DataFrame()
    for cluster in clustered_df['Cluster'].unique():
        cluster_data = clustered_df[clustered_df['Cluster'] == cluster]
        sampled_cluster_data = cluster_data.sample(min(num_samples_per_cluster, len(cluster_data)), random_state=42)
        sampled_data = pd.concat([sampled_data, sampled_cluster_data])

    return sampled_data


num_samples_per_cluster = 10


sampled_data = clustering_sampling(data_df, num_samples_per_cluster)

print(sampled_data.head())
```

```python
sampling_4=[]
X_sampled = sampled_data.drop(target_variable, axis=1)
y_sampled = sampled_data[target_variable]

X_train_sampled, X_test_sampled, y_train_sampled, y_test_sampled = train_test_split(
    X_sampled, y_sampled, test_size=0.2, random_state=42
)


for model_name, model in models.items():
   
    model.fit(X_train_sampled, y_train_sampled)
    
   
    test_predictions = model.predict(X_test_sampled)
    test_accuracy = accuracy_score(y_test_sampled, test_predictions)
    sampling_4.append(test_accuracy)
   
```

```python
answer['Sampling4']=sampling_4
answer
```

### Bootstrap Sampling

```python
target_variable = 'Class'


X = data_df.drop(target_variable, axis=1)
y = data_df[target_variable]

num_bootstrap_samples = 400


bootstrap_samples = pd.DataFrame()


for _ in range(num_bootstrap_samples):
    
    bootstrap_sample_indices = np.random.choice(len(data_df), len(data_df), replace=True)
    bootstrap_sample = data_df.iloc[bootstrap_sample_indices]
    
    # Append the bootstrap sample to the DataFrame
    bootstrap_samples = pd.concat([bootstrap_samples, bootstrap_sample])
```

```python
bootstrap_samples.head()
```

```python
sampling_5=[]
X_bootstrap = bootstrap_samples.drop(target_variable, axis=1)
y_bootstrap = bootstrap_samples[target_variable]


X_train_bootstrap, X_test_bootstrap, y_train_bootstrap, y_test_bootstrap = train_test_split(
    X_bootstrap, y_bootstrap, test_size=0.2, random_state=42
)

for model_name,model in models.items():
    model.fit(X_train_bootstrap, y_train_bootstrap)
    test_predictions = model.predict(X_test_bootstrap)
    test_accuracy = accuracy_score(y_test_bootstrap, test_predictions)
    sampling_5.append(test_accuracy)
```

```python
answer['Sampling5']=sampling_5
```
